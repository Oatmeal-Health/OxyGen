```python
# Sample script to produce a split into train/test/dev
```

```python
import pandas as pd
import random
```

```python
FILENAME = 'filename'

# This file is the result of "ls nlst-tensors/Tensors/ > actual_tensors.txt"
df = pd.read_csv('actual_tensors.txt', header=None)
df.columns = [FILENAME]
df.head()
```

```python
PATIENT = 'patient'
df[PATIENT] = df[FILENAME].apply(lambda x: x.split('__')[2])
df.head()
```

```python
patients = df[PATIENT].unique()
len(patients)
```

```python
random.seed(44)
random.shuffle(patients)

train_dev, test = 0.8, 0.2
train_dev_list = patients[:int(train_dev*len(patients))]
test_list = patients[len(train_dev_list):]

train, dev = 0.88, 0.12
train_list = train_dev_list[:int(train*len(train_dev_list))]
dev_list = train_dev_list[len(train_list):]

print(len(train_list), len(test_list), len(dev_list))
```

```python
SPLIT = 'split'
df_train = pd.DataFrame({PATIENT: train_list, SPLIT: ['train']*len(train_list)})
df_test = pd.DataFrame({PATIENT: test_list, SPLIT: ['test']*len(test_list)})
df_dev = pd.DataFrame({PATIENT: dev_list, SPLIT: ['dev']*len(dev_list)})
```

```python
dff = pd.concat([df_train, df_test, df_dev])
dfm = pd.merge(df, dff)
```

```python
dfm.head(50)
```

```python
for split in ['train', 'test', 'dev']:
    dfm[dfm[SPLIT] == split][[FILENAME]].to_csv(f'{split}.csv', index=False, header=None)
```
